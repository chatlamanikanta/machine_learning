## Problem Statement


You are a data scientist working for a sports analytics company. The dataset, named **"sports_league_data.csv"** includes a variety of attributes for players in the league. The columns in the dataset are as follows:

- **player_id:** A unique identifier for each player.
- **age:** Age of the player.
- **position:** The position the player typically plays in (e.g., Forward, Midfielder, Defender, Goalkeeper).
- **team:** The team the player belongs to.
- **games_played:** The total number of games played by the player in the season.
- **player_rating:** An overall performance rating for the player for the season.

**Import Necessary Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

## Task1

1. Import the data from the "sports_league_data.csv" file.
2. display the number of rows and columns. 
3. Display the first few rows of the dataset to get an overview.


```python
df=pd.read_csv("sports_league_data.csv")
print(df.shape)
df.head(5)
```

## Task2:

**Age Analysis:**
- Mean: Calculate the average age of players in the league.
- Standard Deviation Intervals:
    - Calculate the age range within one standard deviation of the mean.
    - Calculate the age range within two standard deviations of the mean.
    - Calculate the age range within three standard deviations of the mean.
  

```python
avg_age=round(df["age"].mean(),2)
std_age=round(df["age"].std(),2)

l1=round(avg_age-std_age,2)
r1=round(avg_age+std_age,2)

l2=round(avg_age-2*std_age,2)
r2=round(avg_age+2*std_age,2)

l3=round(avg_age-3*std_age,2)
r3=round(avg_age+3*std_age,2)

print(l1,r1)
print(l2,r2)
print(l3,r3)
```

## Task3:

**Player Rating Analysis:**
- Mean: Determine the average player rating across the league.
- Standard Deviation Intervals:
    - Calculate the player rating range within one standard deviation of the mean.
    - Calculate the player rating range within two standard deviations of the mean.
    - Calculate the player rating range within three standard deviations of the mean.

```python
avg_plyrat=round(df["player_rating"].mean(),2)
std_plyrat=round(df["player_rating"].std(),2)

pl1=round(avg_plyrat-std_plyrat,2)
pr1=round(avg_plyrat+std_plyrat,2)

pl2=round(avg_plyrat-2*std_plyrat,2)
pr2=round(avg_plyrat+2*std_plyrat,2)

pl3=round(avg_plyrat-3*std_plyrat,2)
pr3=round(avg_plyrat+3*std_plyrat,2)

print(pl1,pr1)
print(pl2,pr2)
print(pl3,pr3)
```

## Task4:

Create histogram with Kernel Density Estimation (KDE) plots for both the 'age' and 'player_rating' columns from the dataset. While constructing these plots, ensure to visually represent the standard deviation intervals on the graph.


- **'age'**

```python
sb.histplot(df["age"],kde=True)

plt.axvline(avg_age,color='black',linestyle='dashed',linewidth=1)

plt.axvline(l1,color='red',linestyle='dashed',linewidth=1)
plt.axvline(r1,color='red',linestyle='dashed',linewidth=1)
plt.axvline(l2,color='blue',linestyle='dashed',linewidth=1)
plt.axvline(r2,color='blue',linestyle='dashed',linewidth=1)
plt.axvline(l3,color='green',linestyle='dashed',linewidth=1)
plt.axvline(r3,color='green',linestyle='dashed',linewidth=1)

plt.show()
```

- **'player_rating'**

```python
sb.histplot(df["player_rating"],kde=True)

plt.axvline(avg_plyrat,color='black',linestyle='dashed',linewidth=1)

plt.axvline(pl1,color='red',linestyle='dashed',linewidth=1)
plt.axvline(pr1,color='red',linestyle='dashed',linewidth=1)
plt.axvline(pl2,color='blue',linestyle='dashed',linewidth=1)
plt.axvline(pr2,color='blue',linestyle='dashed',linewidth=1)
plt.axvline(pl3,color='green',linestyle='dashed',linewidth=1)
plt.axvline(pr3,color='green',linestyle='dashed',linewidth=1)

plt.show()
```

## Task5

- Calculate the lower and upper boundaries for the 'Age' and 'Player_Rating' columns. 
- For each column, identify values that fall outside the calculated lower and upper boundaries. These values are considered outliers.

- **Display age outliers**

```python
df_age_outliers=df[(df["age"]<l3) | (df["age"]>r3)]
df_age_outliers.shape
```

- **Display player_rating outliers**

```python
df_plyrat_outliers=df[(df["player_rating"]<pl3) | (df["player_rating"]>pr3)]
df_plyrat_outliers.shape
```

## Task6

-  Remove outliers from both the 'age' and 'player_rating' columns and then visualize the adjusted data using histogram plots.


```python
df=df[(df["age"]>l3) & (df["age"]<r3)]
df=df[(df["player_rating"]>pl3) & (df["player_rating"]<pr3)]

```

- **plot the histograms**

```python
sb.histplot(df["age"],kde=True)
plt.show()
```

```python
sb.histplot(df["player_rating"],kde=True)
plt.show()
```
