```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_csv("sports_league_data.csv")
print(df.shape)
df.head(5)
```

```python
mean=round(df["age"].mean(),2)
std=round(df["age"].std(),2)

l1=round(mean-std,2)
r1=round(mean+std,2)
l2=round(mean-2*std,2)
r2=round(mean+2*std,2)
l3=round(mean-3*std,2)
r3=round(mean+3*std,2)

print(l1,r1)
print(l2,r2)
print(l3,r3)
```

```python
plymn=round(df["player_rating"].mean(),2)
plystd=round(df["player_rating"].std(),2)

pl1=round(plymn-plystd,2)
pr1=round(plymn+plystd,2)
pl2=round(plymn-2*plystd,2)
pr2=round(plymn+2*plystd,2)
pl3=round(plymn-3*plystd,2)
pr3=round(plymn+3*plystd,2)

print(pl1,pr1)
print(pl2,pr2)
print(pl3,pr3)
```

```python
def histplot(col,mean,std):
    sb.histplot(col,kde=True)
    plt.axvline(mean,color="black",linewidth=1,linestyle="dashed")
    
    plt.axvline(mean-std,color="red",linewidth=1,linestyle="dashed")
    plt.axvline(mean+std,color="red",linewidth=1,linestyle="dashed")
    plt.axvline(mean-2*std,color="blue",linewidth=1,linestyle="dashed")
    plt.axvline(mean+2*std,color="blue",linewidth=1,linestyle="dashed")
    plt.axvline(mean-3*std,color="green",linewidth=1,linestyle="dashed")
    plt.axvline(mean+3*std,color="green",linewidth=1,linestyle="dashed")
    
    plt.show()
histplot(df["age"],mean,std)
```

```python
histplot(df["player_rating"],plymn,plystd)
```

```python
df1_outliers=df[(df["age"]<l3) | (df["age"]>r3)]
```

```python
df1_outliers.shape
```

```python
df1_no_outliers=df[(df["age"]>l3) & (df["age"]<r3)]
```

```python
histplot(df1_no_outliers["age"],mean,std)
```

```python
df2_no_outliers=df[(df["player_rating"]>pl3) & (df["player_rating"]<pr3)]
```

```python
histplot(df2_no_outliers["player_rating"],plymn,plystd)
```
