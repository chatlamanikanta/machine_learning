```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("heights.csv")
df.head(3)
```

```python
mean=round(df["height"].mean(),2)
std=round(df["height"].std(),2)
```

```python
mean,std
```

```python
sb.histplot(df["height"],kde=True)

plt.axvline(mean,color="black",linewidth=1,linestyle="dashed")

plt.axvline(mean-std, color="red",linewidth=1,linestyle="dashed")
plt.axvline(mean+std, color="red",linewidth=1,linestyle="dashed")
plt.axvline(mean-2*std, color="blue",linewidth=1,linestyle="dashed")
plt.axvline(mean+2*std, color="blue",linewidth=1,linestyle="dashed")

plt.show()
```

```python
l1=round(mean-std,2)
r1=round(mean+std,2)
l1,r1
```

```python
l2=round(mean-2*std,2)
r2=round(mean+2*std,2)
l2,r2
```

```python
l3=round(mean-3*std,2)
r3=round(mean+3*std,2)
l3,r3
```

```python
df_68=df[(df["height"]>l1) & (df["height"]<r1)]
df_68.shape
```

```python
df_95=df[(df["height"]>l2) & (df["height"]<r2)]
df_95.shape
```

```python
df_99=df[(df["height"]>l3) & (df["height"]<r3)]
df_99.shape
```
