```python
import pandas as pd

df_sales = pd.read_excel("linechart.xlsx")
df_sales.head()
```

## Line Chart

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(df_sales['Quarter'],df_sales['Fridge'],color='black',marker='o',label='Fridge')
plt.plot(df_sales['Quarter'],df_sales['Dishwasher'],label='Dishwasher')
plt.plot(df_sales['Quarter'],df_sales['Washing Machine'],label='Washing Machine')
plt.title('Product Sales')
plt.ylabel('Revenue (mln $)')
plt.xlabel('Financial Quarter')
plt.legend()
plt.show()
```

## Pie Chart

```python
total_sales=df_sales[['Fridge','Dishwasher','Washing Machine']].sum()
print(total_sales.index)
total_sales
```

```python
plt.pie(total_sales,labels=total_sales.index,autopct="%1.1f%%",explode=(0.1,0,0),startangle=150,shadow=True)

plt.show()
```

## Bar Chart

```python
df_sales.plot(kind="bar",x='Quarter')
plt.xticks(rotation=360)
plt.show()
```

To use Quarters as X axis, we can set them as an index on a dataframe and then use plot function

```python
df_sales_2=df_sales.set_index('Quarter')
df_sales
```

```python
df_sales_2.loc['Q4 2022']
```

```python
df_sales_2.plot(kind='bar')
plt.xticks(rotation=45)
plt.show()
```

```python
df_sales_2.plot(kind="barh")
plt.show()
```

## Histogram

```python
df_score=pd.read_excel('histograms.xlsx')
df_score.head()
```

```python
plt.hist(df_score['Exam_Score'],density=True)
plt.show()
```

```python
import seaborn as sb
sb.histplot(df_score['Exam_Score'],kde=True)
# or
# sb.histplot(data=df_score,x='Exam_Score',kde=True)

plt.xlabel("Score")
plt.ylabel("Count")
plt.title("Exam Score")

plt.show()
```
