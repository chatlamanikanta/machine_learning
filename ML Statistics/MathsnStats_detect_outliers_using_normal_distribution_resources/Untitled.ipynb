```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heights_with_outliers.csv")
df.head(5)
```

```python
mean=round(df["height"].mean(),2)
std=round(df["height"].std(),2)
mean,std
```

```python
def hist():
    plt.figure(figsize=(12,5))
    sb.histplot(df["height"],kde=True)
    
    plt.axvline(mean,color="black",linestyle="dashed",linewidth=1)
    plt.axvline(mean-std,color="red",linestyle="dashed",linewidth=1)
    plt.axvline(mean+std,color="red",linestyle="dashed",linewidth=1)
    plt.axvline(mean-2*std,color="blue",linestyle="dashed",linewidth=1)
    plt.axvline(mean+2*std,color="blue",linestyle="dashed",linewidth=1)
    plt.axvline(mean-3*std,color="green",linestyle="dashed",linewidth=1)
    plt.axvline(mean+3*std,color="green",linestyle="dashed",linewidth=1)
    
    plt.show()

hist()
```

```python
lower=round(mean-3*std,2)
upper=round(mean+3*std,2)
lower,upper
```

```python
df[(df["height"]<lower) | (df["height"]>upper)]
```

```python
# replace the outliers using median or you can remove them but that is based on the situation
med=df["height"].median()
df["height"]=df["height"].apply(lambda x: med if x<lower or x>upper else x)
```

```python
# no outliers

df[(df["height"]<lower) | (df["height"]>upper)]
```
