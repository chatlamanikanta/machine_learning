```python
import numpy as np

def gd(x,y,costs,lr=0.01,epochs=10000):
    m,b=0.0,0.0

    for epoch in range(epochs):
        y_pred=m*x+b
        error=y-y_pred
        cost=np.mean(error**2)

        dm=-2*np.mean(x*error)
        db=-2*np.mean(error)

        m=m-dm*lr
        b=b-db*lr

        costs.append(cost)

        if epoch%1000==0:
            print(f"m : {m}, b :{b}, epoch : {epoch}, cost : {cost}")
    
    return m,b,m*x



if __name__=="__main__":
    x = np.array([1, 2, 3, 4, 5])
    y = np.array([5, 7, 9, 11, 13])
    costs = []

    m, b, y_pred = gd(x, y, costs)

    # Cost plot
    plt.figure(figsize=(10, 4))
    plt.plot(costs)
    plt.xlabel("Epochs")
    plt.ylabel("Cost")
    plt.title("Cost Function over Epochs")
    plt.grid(True)
    plt.show()

    # Scatter plot of actual vs predicted
    plt.figure()
    plt.scatter(x, y, color='blue', label='Actual y')
    plt.scatter(x, y_pred, color='red', label='Predicted y')
    plt.plot(x, y, color='green', linestyle='--', label='Prediction Line')
    plt.xlabel("x")
    plt.ylabel("y / y_pred")
    plt.title("Actual vs Predicted")
    plt.legend()
    plt.grid(True)
    plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 100])  # with outlier

model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(x)

plt.scatter(x, y, color='red', label='Data points')
plt.plot(x, y_pred, color='blue', label='Best fit line (MSE)')
plt.title("Effect of Outlier on Best Fit Line")
plt.legend()
plt.show()
```
