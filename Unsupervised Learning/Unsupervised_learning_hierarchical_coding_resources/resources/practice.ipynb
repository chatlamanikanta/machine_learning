```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(
    n_samples = 1000,
    noise = 0.05,
    factor = 0.5,
    random_state = 42
) 
print(X)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(X[:,0],X[:,1])
plt.title('Noisy Circles')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2,n_init="auto")
km.fit(X)
km.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=km.labels_,edgecolors="k")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=2, linkage="single")
hc.fit(X)
hc.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=hc.labels_,edgecolors="k")
plt.show()
```
